```python
# Chicago'nun gerçek coğrafi kutusunun (ufak bir esneklik payıyla) dışında kalanlar
aykiri_koordinatlar = df[
    (df["Longitude"] < -88.00) | 
    ((df["Longitude"] > -87.50) & (df["Longitude"] != -1)) | 
    (df["Latitude"] > 42.05) | 
    ((df["Latitude"] < 41.60) & (df["Latitude"] != -1))
]

print(f"Gerçekten dışarıda kalan anlamsız veri sayısı: {len(aykiri_koordinatlar)}")
```
